```python
#Metode Naive  bayes
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
df = pd.read_csv('DataSet/survey lung cancer.csv')
```

```python
#melihat info dataset
df.info()


#melihat info dataset pada 5 baris pertama
df.head()
```

```python
#Mengganti gender dan lung_cancer
df["LUNG_CANCER"].replace(["YES","NO"], [0,1], inplace=True)
df["GENDER"].replace(["M","F"], [0,1], inplace=True)


#Memisahkan dataset menjadi data latihan dan data uji
X = df.drop('LUNG_CANCER', axis=1)
Y = df['LUNG_CANCER']
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
#Melatih model
dt = GaussianNB()
dt.fit(X_train, Y_train)


#Membuat prediksi
Y_pred = dt.predict(X_test)
```

```python
#Mengkalkulasikan akurasi
accuracy = accuracy_score(Y_test, Y_pred)
print("Accuracy:", accuracy)
#Check jenis model yang digunakan
print(type(dt))
```
