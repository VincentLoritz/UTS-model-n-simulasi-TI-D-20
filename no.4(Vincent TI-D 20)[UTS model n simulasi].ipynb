```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

#membaca dataset
df = pd.read_csv('DataSet/Salary_Data.csv')


df.info()
```

```python
#mengekstrak variabel independen (X) dan variabel dependen (Y)
X = df['YearsExperience'].values.reshape(-1,1)
Y = df['Salary'].values.reshape(-1,1)
```

```python
#membuat objek linear regression
model = LinearRegression()

#menyesuaikan model dengan data
model.fit(X, Y)

#memprediksi nilai Y berdasarkan X
Y_pred = model.predict(X)
```

```python
#membuat persebaran data
plt.scatter(X, Y)

#memplot garis regression
plt.plot(X, Y_pred, color='red')

#mengatur label sumbu dan judul
plt.xlabel('YearsExperience')
plt.ylabel('Salary')
plt.title('Linear Regression')


#menampilkan plot dalam bentuk grafik dengan garis regression nya
plt.show()
```
