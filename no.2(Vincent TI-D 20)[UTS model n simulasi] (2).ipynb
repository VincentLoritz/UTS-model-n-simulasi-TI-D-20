```python
import pandas as pd

datasets = pd.read_csv('./survey lung cancer.csv')
```

```python
# Melihat informasi dataset
datasets.info()

# melihat informasi dataset pada 5 baris pertama
datasets.head()
```

```python
datasets["LUNG_CANCER"].replace(["YES","NO"], ["LUNG CANCER","NOT LUNG CANCER"], inplace=True)
datasets["GENDER"].replace(["M","F"], [0,1], inplace=True)

# memisahkan atribut dan label
x = datasets[["GENDER", "AGE", "SMOKING", "YELLOW_FINGERS", "ANXIETY", "PEER_PRESSURE", "CHRONIC DISEASE", "WHEEZING", "ALCOHOL CONSUMING", "COUGHING", "SHORTNESS OF BREATH", "SWALLOWING DIFFICULTY", "CHEST PAIN"]]
y = datasets["LUNG_CANCER"]

# membagi dataset menjadi data latih & data uji
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2,random_state=123)
```

```python
from sklearn.tree import DecisionTreeClassifier

# membuat model DecisionTree
tree_model = DecisionTreeClassifier()

# Melatih model dengan menggunakan data latih
tree_model = tree_model.fit(x_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report

y_pred = tree_model.predict(x_test)

acc_score = round(accuracy_score(y_pred,y_test),3)

print('Accuracy: ', acc_score)
print('classification_report')
print(classification_report(y_pred,y_test))
```

```python
from sklearn.tree import export_graphviz
export_graphviz(
    tree_model,
    out_file="lung_cancer_tree.dot",
    feature_names=["GENDER", "AGE", "SMOKING", "YELLOW_FINGERS", "ANXIETY", "PEER_PRESSURE", "CHRONIC DISEASE", "WHEEZING", "ALCOHOL CONSUMING", "COUGHING", "SHORTNESS OF BREATH", "SWALLOWING DIFFICULTY", "CHEST PAIN"],
    class_names=["LUNG CANCER", "NOT LUNG CANCER"],
    rounded=True,
    filled=True
)
```
